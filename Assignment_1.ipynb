## 3.迴圈:請將程式碼以及結果貼在下方

(a)請使用迴圈，計算出2的幾次方會大於1000000。(20%)

```python
a = 1
b = 0
while a <= 10000000 :
    a = a*2
    b = b + 1
ans = b
print(ans)
```

(b)請使用迴圈，計算出n!大於500000。(!代表階層，ex:5!=1x2x3x4x5)(20%)

```python
a = 1
b = 0
while a <= 500000 :
    b = b + 1
    a = a * b
ans = b
print(ans)
```
